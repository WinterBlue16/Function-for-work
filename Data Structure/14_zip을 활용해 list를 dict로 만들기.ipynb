```python
# 서로 연관이 있거나 연결된 데이터를 다룰 때, dict는 매우 유용하게 쓰인다. 
# 데이터 프레임의 경우 서로 다른 열을 key, value로 하여 dict를 만들고 싶을 때가 있다. 

# 라이브러리 불러오기
import pandas as pd

# 데이터 불러오기
# 이전에 크롤링해둔 데이터를 사용한다.
sample_df = pd.read_csv('postype_crawling.csv', encoding='cp949', index_col=0)
sample_df
```

```python
# 글 제목-작가 링크를 dict로 만든다
# list로 만들기
title_li = list(sample_df['title'])
writer_li= list(sample_df['writer'])

# dict로 만들기
tw_dict = dict(zip(title_li, writer_li))
tw_dict
```
