# 데이터프레임 index 리셋하기
> 데이터프레임을 다루다 보면 결측치 혹은 이상치를 제거하고, 필요한 데이터만 인덱싱하는 과정에서 인덱스 번호가 뒤섞이게 된다. 만약 새롭게 인덱스 번호를 매기고 싶을 때, reset을 사용할 수 있다.

효과적인 시연을 위해 kaggle의 [Crimes in Chicago](https://www.kaggle.com/currie32/crimes-in-chicago)를 가져왔다.

```python
# 라이브러리 불러오기
import pandas as pd
```

```python
# 샘플 데이터 가져오기
df = pd.read_csv('archive\Chicago_Crimes_2012_to_2017.csv')
df.head(10)
```

```python
# 데이터 형태 확인
df.info()
```

```python
# '범죄 유형'에 따른 데이터 추출
df['Primary Type'].value_counts()
```

```python
# KIDNAPPING
kidnap_df = df[df['Primary Type']=='KIDNAPPING']

# HUMAN TRAFFICKING 
HT_df = df[df['Primary Type']=='HUMAN TRAFFICKING']
```

```python
kidnap_df
```

```python
HT_df
```

```python
# INDEX reset
kidnap_df = kidnap_df.reset_index(drop=True)
kidnap_df
```

```python
HT_df = HT_df.reset_index(drop=True)
HT_df
```

```python
# drop = False 옵션의 경우
# ARSON
arson_df = df[df['Primary Type']=='ARSON']
arson_df = arson_df.reset_index(drop=False)
arson_df # 기존의 index 열을 그대로 남겨둔 채 가장 왼쪽에 새로운 index 열이 생긴다!
```
