# 03. OpenCV_Matplotlib.pyplot
> 이미지를 새창으로 여는 것이 아니라 주피터 노트북에 직접 출력하고 싶을 때 이용하는 함수이다. OpenCV 외에도 시각화 함수로써 유용하게 사용되고 있다. 

```python
# 라이브러리 불러오기
import cv2 
import matplotlib.pyplot as plt

# 이미지 파일 불러오기
img = cv2.imread('unnamed.png')
img
```

```python
# 이미지 새창 열지 않고 출력하기
plt.imshow(img)
```

```python
# x, y축 없이 이미지 출력해보기
plt.imshow(img)
plt.axis('off')
```

보다시피 이미지가 성공적으로 주피터 노트북에서 출력되는 것을 볼 수 있다. 이번엔 다른 이미지를 출력해볼까?

```python
# 다른 이미지 불러오기
img2 = cv2.imread("summerfield-336672_1920_resize.jpg")

# 새창으로 이미지 출력하기
cv2.imshow('image2', img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 이미지 출력하기
plt.imshow(img2)
```

새창에서 출력된 이미지와 주피터 노트북에 출력된 이미지가 다르다! 햇살 탓에 오렌지와 붉은 빛이 돌았던 원본 이미지와 달리 노트북에 출력된 이미지는 푸른 마스크가 씌워진 것처럼 보인다. 어째서일까?
